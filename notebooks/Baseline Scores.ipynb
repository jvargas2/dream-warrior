```python
def plot(name, time, rewards, losses):
    plt.figure(figsize=(20,5))
    plt.subplot(131)
    plt.title('%s Rewards (Mean %.2f)' % (name, mean(rewards)))
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    plt.plot(rewards)
    plt.subplot(132)
    plt.title('%s Losses (Mean %.4f)' % (name, mean(losses)))
    plt.ylabel('Mean Loss')
    plt.plot(losses)
    plt.show()
    
    print('Time Elapsed: %s' % time)
    print('Episodes: %d' % len(rewards))
    
    max_rewards = {}
    rewards_copy = rewards.copy()
    
    for _ in range(5):
        max_reward = max(rewards_copy)
        max_index = rewards.index(max_reward)
        rewards_copy.remove(max_reward)
        max_rewards[max_index] = max_reward

    print('Highest Rewards:')
    
    for key, value in max_rewards.items():
        print('  Episode %d: %f' % (key, value))
```
