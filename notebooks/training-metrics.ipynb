```python
# Imports
import numpy as np
from numpy.polynomial.polynomial import polyfit
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def plot_rewards(name, rewards):
    episodes = np.arange(len(rewards))

    b, m = polyfit(episodes, rewards, 1)

    fig = plt.figure()
    fig.suptitle(name + ' Training Rewards')
    plt.xlabel('Episode')
    plt.ylabel('Reward')
    
    plt.plot(episodes, rewards, '.')
    plt.plot(episodes, b + m * episodes, '-')
    
    print('Episodes: %d' % len(rewards))
    print('Max Reward: %d (Episode: %d)' % (max(rewards), rewards.index(max(rewards))))
```
